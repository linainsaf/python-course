```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesAnalysis:
    
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path, encoding='latin-1')

    def clean_data(self):
        # Remove duplicates
        self.df.drop_duplicates(inplace=True)

        # Fill missing values with 0
        self.df.fillna(0, inplace=True)

        # Convert Order Date to datetime object
        self.df['Order Date'] = pd.to_datetime(self.df['Order Date'])

        # Convert Quantity to int
        self.df['Quantity'] = self.df['Quantity'].astype(int)

        # Convert Sales to float
        self.df['Sales'] = self.df['Sales'].astype(float)

    def total_revenue_by_year_region(self):
        return self.df.groupby(['Order Date', 'Region'])['Sales'].sum().reset_index()

    def highest_sales_revenue(self):
        return self.df.groupby(['Order Date', 'Region'])['Sales'].sum().idxmax()

    def average_sales_revenue_per_month(self):
        return self.df.groupby(['Region', pd.Grouper(key='Order Date', freq='M')])['Sales'].mean().reset_index()

    def highest_sales_category(self):
        return self.df.groupby(['Category'])['Sales'].sum().idxmax()

    def percentage_contribution(self):
        return self.df.groupby(['Category'])['Sales'].sum() / self.df['Sales'].sum() * 100

    def visualize_total_revenue(self):
        total_revenue = self.total_revenue_by_year_region()
        plt.figure(figsize=(12, 8))
        sns.barplot(x='Order Date', y='Sales', hue='Region', data=total_revenue)
        plt.title('Total Revenue by Year and Region')
        plt.xlabel('Year')
        plt.ylabel('Revenue')
        plt.show()

    def visualize_average_sales_revenue(self):
        average_sales = self.average_sales_revenue_per_month()
        plt.figure(figsize=(12, 8))
        sns.lineplot(x='Order Date', y='Sales', hue='Region', data=average_sales)
        plt.title('Monthly Average Sales Revenue by Region')
        plt.xlabel('Month')
        plt.ylabel('Average Sales Revenue')
        plt.show()

    def visualize_percentage_contribution(self):
        percentage = self.percentage_contribution()
        plt.figure(figsize=(12, 8))
        plt.pie(percentage, labels=percentage.index, autopct='%1.1f%%')
        plt.title('Percentage Contribution of Each Product Category to Total Revenue')
        plt.show()

    def statistical_analysis(self):
        return self.df.groupby(['Region'])['Sales'].agg([np.mean, np.median, np.std])
    
    def generate_report(self):
        print("\nSALES REPORT")
        print("============")
        print("Total Revenue by Year and Region:\n", self.total_revenue_by_year_region())
        print("\nYear and Region with Highest Sales Revenue:\n", self.highest_sales_revenue())
        print("\nAverage Sales Revenue per Month by Region:\n", self.average_sales_revenue_per_month())
        print("\nProduct Category with Highest Sales Revenue:\n", self.highest_sales_category())
        print("\nPercentage Contribution of Product Categories to Total Revenue:\n", self.percentage_contribution())
        
        #self.visualize_total_revenue()
        self.visualize_average_sales_revenue()
        self.visualize_percentage_contribution()
        
        print("\nStatistical Analysis of Sales Revenue by Region:")
        sales_stats = self.statistical_analysis()
        for index, row in sales_stats.iterrows():
            region = index
            print(f"\n{region}:")
            print("Mean:", row["mean"])
            print("Median:", row["median"])
            print("Standard Deviation:", row["std"])
        
    def save_clean_data(self, file_path):
        self.df.to_csv(file_path, index=False)

    def save_figure(self, fig, file_path):
        fig.savefig(file_path)
```

```python
class SalesAnalysis:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path, encoding='latin-1')
    
    def clean_data(self, drop_duplicates=True, fillna=True, date_format='%m/%d/%Y', 
                   convert_types={'Order Date': 'datetime64[ns]', 'Quantity': int, 'Sales': float}):
        """
        This method cleans the data by handling missing values, duplicates, and converting data types.
        """
        if drop_duplicates:
            self.df.drop_duplicates(inplace=True)
        if fillna:
            # Fill missing values with 0
            self.df.fillna(0, inplace=True)
        self.df = self.df.astype(convert_types)
        self.df['Order Date'] = pd.to_datetime(self.df['Order Date'], format=date_format)
    
    def filter_by_date(self, start_date=None, end_date=None):
        """
        This method filters the data by date, keeping only rows where the Order Date column is between start_date and end_date.
        """
        if start_date is not None:
            self.df = self.df[self.df['Order Date'] >= pd.to_datetime(start_date)]
        if end_date is not None:
            self.df = self.df[self.df['Order Date'] <= pd.to_datetime(end_date)]
            
    def sales_by_category(self, category=None):
        if category is None:
            sales_data = self.df.groupby('Category')['Sales'].sum()
        else:
            sales_data = self.df[self.df['Category'] == category].groupby('Category')['Sales'].sum()
        return sales_data
    
    def sales_by_region(self, region=None):
        if region is None:
            sales_data = self.df.groupby('Region')['Sales'].sum()
        else:
            sales_data = self.df[self.df['Region'] == region].groupby('Region')['Sales'].sum()
        return sales_data
    
    def sales_by_subcategory(self, subcategory=None):
        if subcategory is None:
            sales_data = self.df.groupby('Sub-Category')['Sales'].sum()
        else:
            sales_data = self.df[self.df['Sub-Category'] == subcategory].groupby('Sub-Category')['Sales'].sum()
        return sales_data
    
    def quantity_by_category(self, category=None):
        if category is None:
            quantity_data = self.df.groupby('Category')['Quantity'].sum()
        else:
            quantity_data = self.df[self.df['Category'] == category].groupby('Category')['Quantity'].sum()
        return quantity_data
    
    def shipping_cost_by_region(self, region=None):
        if region is None:
            shipping_data = self.df.groupby('Region')['Shipping Cost'].sum()
        else:
            shipping_data = self.df[self.df['Region'] == region].groupby('Region')['Shipping Cost'].sum()
        return shipping_data
    
    def discount_by_subcategory(self, subcategory=None):
        if subcategory is None:
            discount_data = self.df.groupby('Sub-Category')['Discount'].mean()
        else:
            discount_data = self.df[self.df['Sub-Category'] == subcategory].groupby('Sub-Category')['Discount'].mean()
        return discount_data
    
    def total_sales(self, category=None, region=None):
        if category and region:
            return self.df[(self.df['Category'] == category) & (self.df['Region'] == region)]['Sales'].sum()
        elif category:
            return self.df[self.df['Category'] == category]['Sales'].sum()
        elif region:
            return self.df[self.df['Region'] == region]['Sales'].sum()
        else:
            return self.df['Sales'].sum()
        
    def total_profit(self, category=None, region=None):
        if category and region:
            return self.df[(self.df['Category'] == category) & (self.df['Region'] == region)]['Profit'].sum()
        elif category:
            return self.df[self.df['Category'] == category]['Profit'].sum()
        elif region:
            return self.df[self.df['Region'] == region]['Profit'].sum()
        else:
            return self.df['Profit'].sum()
        
    def average_sales(self, category=None, region=None):
        if category and region:
            return self.df[(self.df['Category'] == category) & (self.df['Region'] == region)]['Sales'].mean()
        elif category:
            return self.df[self.df['Category'] == category]['Sales'].mean()
        elif region:
            return self.df[self.df['Region'] == region]['Sales'].mean()
        else:
            return self.df['Sales'].mean()
        
    def average_profit(self, category=None, region=None):
        if category and region:
            return self.df[(self.df['Category'] == category) & (self.df['Region'] == region)]['Profit'].mean()
        elif category:
            return self.df[self.df['Category'] == category]['Profit'].mean()
        elif region:
            return self.df[self.df['Region'] == region]['Profit'].mean()
        else:
            return self.df['Profit'].mean()
        
    def top_n_sales_categories(self, n=5):
        df_sales_category = self.df.groupby('Category')['Sales'].sum().reset_index().sort_values('Sales', ascending=False).head(n)
        return df_sales_category
    
    def worst_n_profit_margin_regions(self, n=5):
        df_profit_region = self.df.groupby('Region')['Profit'].sum().reset_index()
        df_sales_region = self.df.groupby('Region')['Sales'].sum().reset_index()
        df_merged = pd.merge(df_profit_region, df_sales_region, on='Region', how='inner')
        df_merged['Profit Margin'] = df_merged['Profit'] / df_merged['Sales'] * 100
        df_worst_n = df_merged.sort_values('Profit Margin', ascending=True).head(n)
        return df_worst_n
    
    def total_revenue_by_year_region(self):
        df_year_region_revenue = self.df.groupby(['Year', 'Region'])['Sales'].sum().reset_index()
        return df_year_region_revenue
    
    def highest_sales_year_region(self):
        df_year_region_revenue = self.df.groupby(['Year', 'Region'])['Sales'].sum().reset_index()
        max_revenue = df_year_region_revenue['Sales'].max()
        year_region = df_year_region_revenue[df_year_region_revenue['Sales'] == max_revenue].iloc[0]
        return year_region
    
    def average_sales_revenue_per_month_region(self):
        df_sales_month_region = self.df.groupby(['Region', pd.Grouper(key='Order Date', freq='M')])['Sales'].sum().reset_index()
        df_average_sales = df_sales_month_region.groupby('Region')['Sales'].mean().reset_index()
        return df_average_sales
    
    def highest_sales_category(self):
        df_sales_category = self.df.groupby('Category')['Sales'].sum().reset_index()
        max_sales = df_sales_category['Sales'].max()
        category = df_sales_category[df_sales_category['Sales'] == max_sales].iloc[0]
        return category
    
    def category_sales_contribution(self):
        df_sales_category = self.df.groupby('Category')['Sales'].sum().reset_index()
        total_sales = df_sales_category['Sales'].sum()
        df_sales_category['Sales Contribution'] = df_sales_category['Sales'] / total_sales * 100
        return df_sales_category
    
    def visualize_sales_by_category(self, category=None):
        data = self.sales_by_category(category)
        sns.barplot(x='Category', y='Sales', data=data)
        plt.show()

    def visualize_sales_by_region(self, region=None):
        data = self.sales_by_region(region)
        sns.barplot(x='Region', y='Sales', data=data)
        plt.show()

    def visualize_sales_by_subcategory(self, category=None):
        data = self.quantity_by_category(category)
        sns.barplot(x='Category', y='Quantity', data=data)
        plt.show()

    def visualize_quantity_by_category(self, category=None):
        data = self.quantity_by_category(category)
        sns.barplot(x='Category', y='Quantity', data=data)
        plt.show()

    def visualize_shipping_cost_by_region(self, region=None):
        data = self.shipping_cost_by_region(region)
        sns.barplot(x='Region', y='Shipping Cost', data=data)
        plt.show()

    def visualize_discount_by_subcategory(self, subcategory=None):
        data = self.discount_by_subcategory(subcategory)
        sns.barplot(x='Sub-Category', y='Discount', data=data)
        plt.show()

    def visualize_yearly_avg_sales_region(self):
        data = self.df.groupby(['Order Year', 'Region'])['Sales'].mean().reset_index()
        sns.lineplot(x='Order Year', y='Sales', hue='Region', data=data)
        plt.show()

    def visualize_category_contribution(self):
        data = self.category_contribution_to_total_sales()
        plt.pie(data['Total Sales'], labels=data['Category'], autopct='%1.1f%%')
        plt.show()
    
    def generate_report(self, start_date=None, end_date=None):
        # filter data by date range if start and/or end date are provided
        if start_date and end_date:
            self.filter_by_date(start_date=start_date, end_date=end_date)
        
        # calculate summary statistics
        total_sales = self.total_sales()
        total_profit = self.total_profit()
        avg_sales = self.average_sales()
        avg_profit = self.average_profit()
        
        # find top sales categories and worst performing regions
        top_categories = self.top_n_sales_categories(n=5)
        worst_regions = self.worst_performing_regions(n=5)
        
        # calculate total revenue by year and region
        revenue_by_year_region = self.total_revenue_by_year_region()
        
        # find year and region with highest sales revenue
        highest_sales_year, highest_sales_region = self.highest_sales_year_region()
        
        # calculate average sales revenue per month by region
        avg_sales_per_month = self.average_sales_revenue_per_month()
        
        # find product category with highest sales revenue
        highest_sales_category = self.highest_sales_category()
        
        # calculate percentage contribution of each product category to total sales revenue
        category_contribution = self.category_contribution()
        
        # visualize data
        self.visualize_sales_by_category()
        self.visualize_sales_by_region()
        self.visualize_sales_by_subcategory()
        self.visualize_quantity_by_category()
        self.visualize_shipping_cost_by_region()
        self.visualize_discount_by_subcategory()
        self.visualize_yearly_avg_sales_region()
        self.visualize_category_contribution()
        
        # Save cleaned data to new CSV file
        self.df.to_csv('cleaned_data.csv', index=False)
        
        # Save figures
        plt.savefig('sales_by_category.png')
        plt.savefig('sales_by_region.png')
        
        # create report
        report = f"Sales Analysis Report\n\nDate Range: {start_date} to {end_date}\n\n"
        report += f"Total Sales: {total_sales}\nTotal Profit: {total_profit}\nAverage Sales: {avg_sales}\nAverage Profit: {avg_profit}\n\n"
        report += f"Top Sales Categories:\n{top_categories}\n\n"
        report += f"Worst Performing Regions:\n{worst_regions}\n\n"
        report += f"Total Revenue by Year and Region:\n{revenue_by_year_region}\n\n"
        report += f"Highest Sales Year and Region: {highest_sales_year}, {highest_sales_region}\n\n"
        report += f"Average Sales Revenue per Month by Region:\n{avg_sales_per_month}\n\n"
        report += f"Highest Sales Category: {highest_sales_category}\n\n"
        report += f"Category Contribution to Total Sales Revenue:\n{category_contribution}"
        
        return report
```
